```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
import pickle
```

```python
df = pd.read_csv('C:/Users/Malcolm/Documents/MedicalExpenditure/data/meps_data_2019_new_feats.csv')
pickle_in = open("C:/Users/Malcolm/Documents/MedicalExpenditure/MEPS_gb_model_non_processed_v1.pickle", "rb")
reg_gb = pickle.load(pickle_in)
```

```python
reg_gb.best_params_
```

```python
# train test split
X = df.drop('total_expenditure', axis =1)
y = df.total_expenditure.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

Here we'll import the params from the GridSearchCV search in gradient_boosting1=.py

```python
gb_1 = GradientBoostingRegressor(max_depth = 6,
 min_samples_leaf = 7,
 min_samples_split = 2,
 n_estimators = 72)
```

```python
gb_1.fit(X_train, y_train)
```

```python
sorted_idx = gb_1.feature_importances_.argsort()
plt.barh(df.columns[sorted_idx][:15], gb_1.feature_importances_[sorted_idx][:15])
plt.xlabel("GradientBoostingRegressor Feature Importance")
```
